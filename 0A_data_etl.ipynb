# Q2 - 2025

```python
from bs4 import BeautifulSoup
import requests

html = requests.get('https://www.insidermonkey.com/blog/the-coca-cola-company-nyseko-q2-2025-earnings-call-transcript-1573661/').text
soup = BeautifulSoup(html, "html.parser")
text = soup.get_text(separator="\n", strip=True)
marker = "Operator:"
idx = text.find(marker)
transcript = text[idx:] if idx != -1 else text
transcript = transcript.replace(":\n",": ")
#transcript = transcript.replace("\n","\n")
print(transcript)

file_path=(r"C:\Git projects\sales_marketing_rag\data\raw\Q2-2025_call.txt")
with open(file_path, "w", encoding="utf-8") as f:
    f.write(transcript)
```
